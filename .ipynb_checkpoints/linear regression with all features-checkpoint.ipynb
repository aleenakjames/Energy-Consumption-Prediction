```python
import pandas as pd
df = pd.read_csv("HomeC.csv")
df = df.drop(df.columns[23], axis=1)
df = df.dropna()
df['time'] = df['time'].astype(float)
numeric_cols = df.select_dtypes(include=['int', 'float']).columns
df = df.drop(columns=df.columns.difference(numeric_cols))
# df = df.head()
df
```

```python
x = df.iloc[:, 1]
x = x.values.reshape(-1, 1)
y = df.iloc[:, [col for col in range(len(df.columns)) if col != 1]]
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 10)
lnr = LinearRegression().fit(x_train, y_train)
y_pred = lnr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f"Accuracy: {r2}")
```
